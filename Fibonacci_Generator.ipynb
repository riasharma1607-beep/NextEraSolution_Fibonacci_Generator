```python
# Task 1: Fibonacci Generator
# NextEra Solution With Technology Internship

def fibonacci_generator(n):
    a, b = 0, 1
    series = []

    for _ in range(n):
        series.append(a)
        a, b = b, a + b

    return series

try:
    num = int(input("Enter the number of Fibonacci terms: "))

    if num <= 0:
        print("Please enter a positive number.")
    else:
        result = fibonacci_generator(num)
        print("Fibonacci Series:")
        print(*result)

except ValueError:
    print("Invalid input! Please enter an integer.")
```
